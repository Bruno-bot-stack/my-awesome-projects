#### Reading dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('50_Startups.csv')
df
```

#### Independent and Dependent variables

```python
#separate the other attributes from the predicting attribute
x = df.drop('Profit',axis=1)
#separte the predicting attribute into Y for model training 
y = df['Profit']
```

```python
x.head()
```

```python
y.head()
```

#### Handling categorical variables

```python
# handle categorical variable
states=pd.get_dummies(x)
# dropping extra column
states= states.drop('R&D Spend', axis=1)
states= states.drop('Administration', axis=1)
states= states.drop('Marketing Spend', axis=1)
states
```

#### Independent and Dependent variables

```python
# dropping extra column
x= x.drop('State', axis=1)
```

```python
# concatation of independent variables and new cateorical variable.
x=pd.concat([x, states], axis=1)
x
```

#### Splitting Data

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

#### Applying model

```python
# importing module
from sklearn.linear_model import LinearRegression
# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train, y_train)
```

```python
y_prediction = LR.predict(x_test)
y_prediction
```

#### r2_score

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print('r2 socre is ',score)
print('mean_sqrd_error is==', mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is==', np.sqrt(mean_squared_error(y_test,y_prediction)))
```
